# K-mean algorithm

```python
# Standardize the Data
#return_history_standard = returnhissory2
#returnhistory1=returnhistory

from sklearn.preprocessing import StandardScaler

return_history_standard = StandardScaler().fit_transform(return_history1.iloc[:,0:])

from sklearn.decomposition import PCA

pca = PCA().fit(return_history_standard)

loadings=pd.DataFrame(pca.components_,
                      index=['p'+str(i) for i in range(0,28)], 
                      columns=return_history1.iloc[:,0:].columns).T
loadings
```

```python
# Lets load the modules first

from sklearn.cluster import KMeans
loadings
```

```python
# Create a grouping variable using kmeans

wcss = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit_predict(loadings.iloc[:,0:2])
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 8), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# 3 is the optimum number
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit_predict(loadings.iloc[:,0:2])

cluster=pd.DataFrame([loadings.index,kmeans.labels_],index=['Asset Class','Group']).T
print(cluster)

loadings.plot(x='p0',y='p1', marker="o" , linestyle='None', figsize=(15,10), label='Asset Class')
for i in loadings.index:
    plt.annotate(i, (loadings.loc[i][0], loadings.loc[i][1]), size=10, xytext=(0,0), ha='right', textcoords='offset points')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')

plt.xlabel('p0')
plt.ylabel('p1')

plt.show()
```

# Factor-Based Risk Parity
After grouping PCA factors into two groups (0, 1) using K-means learning, we construct a two-tier â€œRisk Parityâ€ portfolio.
 Tier 1: seeking equal risk-exposure across PCA Factors.
 Tier 2: diversifying across asset classes with equal risk-exposure

```python
display(cluster)
```

```python
# Two-Tier â€œRisk Parity"

# Asset Volatilities 
ReturnVol = return_history1.std()
display(ReturnVol)

# Risk Parity 
Weight=1/22/ReturnVol[['SPY', 'IWM', 'LQD', 'EEM',  'EWJ', 'SPYG', 'IWD', 'XLK','XLV', 'XLF', 'XLI', 'XLP', 'XLE',  'AGG', 'SHY', 'QQQ', 'JNK', 'DBC', 'GLD', 'OILNF', 'SPYV', 'MTUM']]
Weight=Weight.append(1/6/ReturnVol[['EFA', 'FXI','XLU', 'XLB','XLY','TLT']])

Weight=Weight/np.sum(Weight)
RiskContr=Weight*ReturnVol
pf=pd.DataFrame([ReturnVol,Weight,RiskContr],index=['Vol','Weight','Risk Contribution'])

cluster2=cluster[['Group']].T
cluster2.columns=pf.columns
pf=pf.append(cluster2, ignore_index = True)

display(pf)
```

```python
# Strategy Simulation with Return_history1

StrategyRet=pd.DataFrame(return_history1.iloc[:,0:].dot(Weight),
                         columns=['StrategyRet'])
StrategyRet.index=return_history1.index

((1 + StrategyRet).cumprod() - 1).plot(title='Strategy Cumulative Returns',figsize=(15,10));
```

```python
performance=pd.DataFrame(pow((1 + StrategyRet).cumprod().iloc[-1,:],12/(StrategyRet.shape[0]-1))-1)
performance.columns=[['Annualized Return']]

performance['Annualized volatility']=pd.DataFrame(StrategyRet).std()*math.sqrt((12))
performance
```
